This notebook merges the ['1137190.xlsx', '11455478.xlsx', '11455508.xlsx', '380631122032201.xlsx'] excel workbooks

```python
import pandas as pd
import os
```

```python
filelist = os.listdir()
filelist = [file for file in filelist if file[-5:] == '.xlsx']
filelist
```

```python
df1 = pd.read_excel(filelist[0])
df2 = pd.read_excel(filelist[1])
df3 = pd.read_excel(filelist[2])
df4 = pd.read_excel(filelist[3])
```

```python
df = pd.concat([df1,df2,df3,df4])
```
