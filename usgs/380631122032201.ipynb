The following stationNZ codes are going to be used:

```python
import pandas as pd
import numpy as np
import datetime
```

```python
NZlist = ['NZS42', 'NZ028', 'NZ032']
```

```python
df = pd.read_excel('../usgs/380631122032201.xlsx')
zoo_df = pd.read_excel('../zooplankton/zoo_df.xlsx')
```

```python
df.drop(labels = 0, inplace = True)
```

```python
zoo_df = zoo_df.loc[zoo_df.StationNZ.isin(NZlist)]
```

```python
zoo_df.SampleDate = pd.to_datetime(zoo_df.SampleDate)
df.datetime = pd.to_datetime(df.datetime)
```

```python
from gauss import gauss_dist
dist = gauss_dist(11)
dist
```

```python
df.columns
```

```python
#df.drop(columns = ['site_no'], inplace = True)
```

```python
df.columns = ['datetime', 'organic_matter', 'nitrate_nitrite', 'ph',
              'temp_c', 'specific_conductance', 'turbidity', 'dissolved_o2_mg_l',
              'fChl', 'salinity', 'dissolved_o2_percent_sat']
```

```python
df['11_day_temp_c'] = np.nan
df['11_day_specific_conductance'] = np.nan
df['11_day_ph'] = np.nan
df['11_day_dissolved_o2_mg_l'] = np.nan
df['11_day_turbidity'] = np.nan
df['11_day_nitrate_nitrite'] = np.nan
df['11_day_organic_matter'] = np.nan
df['11_day_salinity'] = np.nan
df['11_day_dissolved_o2_percent_sat'] = np.nan
```

```python
df.tail()
```

```python
df = df.loc[df.datetime >= (zoo_df.SampleDate.min() - datetime.timedelta(days = 12))]
df = df.loc[df.datetime < datetime.datetime(2020,1,1)]

for i in range(11,len(df)):
    df.loc[i,'11_day_temp_c'] = sum(np.multiply(dist,list(df.temp_c[i-11:i])))
    df.loc[i,'11_day_specific_conductance'] = sum(np.multiply(dist,list(df.specific_conductance[i-11:i])))
    df.loc[i,'11_day_ph'] = sum(np.multiply(dist,list(df.ph[i-11:i])))
    df.loc[i,'11_day_dissolved_o2_mg_l'] = sum(np.multiply(dist,list(df.dissolved_o2_mg_l[i-11:i])))
    df.loc[i,'11_day_turbidity'] = sum(np.multiply(dist,list(df.turbidity[i-11:i])))
    df.loc[i,'11_day_nitrate_nitrite'] = sum(np.multiply(dist,list(df.nitrate_nitrite[i-11:i])))
    df.loc[i,'11_day_organic_matter'] = sum(np.multiply(dist,list(df.organic_matter[i-11:i])))
    df.loc[i,'11_day_salinity'] = sum(np.multiply(dist,list(df.salinity[i-11:i])))
    df.loc[i,'11_day_dissolved_o2_percent_sat'] = sum(np.multiply(dist,list(df.dissolved_o2_percent_sat[i-11:i])))

df = df[11:]
```

```python
df
```

```python
zoo_df = zoo_df.merge(df, how='left', left_on='SampleDate', right_on='datetime')
zoo_df
```

```python
zoo_df = zoo_df.loc[zoo_df.SampleDate >= datetime.datetime(2017,10,10)]
```
