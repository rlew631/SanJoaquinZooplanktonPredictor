```python
import pandas as pd
from datetime import datetime
import numpy as np
```

```python
zoo_df = pd.read_excel('Pump CPUE Matrix 1972-2019.xlsx')
```

```python
zoo_df = zoo_df[['SampleDate','StationNZ','Region','TotalLimno','ALLCYCADULTS','ALLCYCJUV','ALLCOPNAUP','ALLROTIFERS']]
```

```python
type(zoo_df['SampleDate'][0])
```

```python
zoo_df = zoo_df.loc[zoo_df.SampleDate > datetime(2010,1,1)]
zoo_df.to_excel('zoo_df.xlsx')
zoo_df.head()
```

```python
station_lookup = pd.read_csv('Station_lookup.csv')
station_lookup.head()
```

```python
station_lookup = station_lookup[['StationNZ', 'Location', 'lat_decimal', 'lon_decimal', 'StartYr', 'EndYr']]
station_lookup = station_lookup.loc[(station_lookup.EndYr < 2010) == False] #drops entries which end before 2010 and allow NaNs
station_lookup = station_lookup.loc[station_lookup.lat_decimal > 1].drop(columns = ['StartYr','EndYr'])
```

```python
station_lookup.head()
```

```python
station_lookup.to_csv('station_lookup_trimmed.csv')
```

```python
new_zoo = zoo_df.merge(station_lookup, how = 'left', left_on = 'StationNZ', right_on = 'StationNZ')
new_zoo.head()
```
